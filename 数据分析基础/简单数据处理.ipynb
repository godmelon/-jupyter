# 日期格式转换，实现日期的加减

```python
#进行日期的转换的时候，指定原始数据日期格式，有利于转换效率
users['date_account_created'] = pd.to_datetime(users['date_account_created'], format='%Y-%m-%d')
```

```python
#现在再查看,发现类型是datetime
users['date_account_created']
```

```python
#成功转换成时间格式之后，就可以进行时间上的计算，计算相差的天数
users.loc[0, 'date_account_created'] - users.loc[1, 'date_account_created']
```

```python
#定义到时分秒的关系，设置format函数参数，针对一些非常规的数据
users["timestamp_first_active"] = pd.to_datetime(users["timestamp_first_active"],format="%Y%m%d%H%M%S")
users["timestamp_first_active"]
```
